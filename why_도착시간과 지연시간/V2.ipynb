```python
# 1. 결측치 확인
print(df.isnull().sum())
```

```python
# 1. 결측치가 있는 행(Row) 제거
df.dropna(subset=['Arrival Delay in Minutes'], inplace=True)
# 2. 잘 지워졌는지 확인 (0이 나와야 함)
print("제거 후 결측치 수:", df['Arrival Delay in Minutes'].isnull().sum())
```

```python
#step1... 지연은 고객의 만족도에 영향을 미치는가?
```

```python
from scipy import stats
group_sat=df[df['satisfaction']=='satisfied']
group_unsat = df[df['satisfaction']=='neutral or dissatisfied']

t_stat_dep, p_val_dep= stats.ttest_ind(group_sat['Departure Delay in Minutes'],
                                       group_unsat['Departure Delay in Minutes'],
                                       equal_var=False)

t_stat_arr, p_val_arr= stats.ttest_ind(group_sat['Arrival Delay in Minutes'],
                                       group_unsat['Arrival Delay in Minutes'],
                                       equal_var=False)

print(f"====1.출발 지연(Depature Delay) 검정====")
print(f"T-static:{t_stat_dep:.4f}")
print(f"p-value:{p_val_dep:.30f}")

print(f"\n=== 2. 도착 지연(Arrival Delay) 검정 ===")
print(f"T-statistic: {t_stat_arr:.4f}")
print(f"P-value: {p_val_arr:.30f}")
```

```python
# H0: 지연은 승객의 만족도와 아무런 영향이 없다
# H1: 지연은 승객의 만족도에 영향을 미친다
# -> 귀무가설 기각, 지연(출발/도착 모두)은 승객의 만족도에 통계적으로 매우 유의미한 영향을 미칩니다

# t값이  |출발지연|16.55 < |도착지연|18.86
#  고객들은 출발이 늦어지는 것보다 도착이 늦어지는 것에 대해 더 민감하게 반응하며, 
# 도착 지연이 만족/불만족을 가르는 **더 핵심적인 변수

# PS.예상질문: T-TEST 진행 이유(이전 교수님 질문 펜타킬!!) z-test, anova, 카이제곱...
```

```python
#step2.. (deep dive): 도착지연에 고객들이 더 민감하게 반응한다.
#그럼 늦게출발했어도 도착은 제때 하면 괜찮은건가?
# departure delay in minutes : x,  arrival delay in minutes:0


#ps. 예상질문: ontime이 왜 5분인가? 0분 아니예여? 제때 도착한거면.



# 귀무가설: 4개 지연 그룹(On-Time, Recovered, Delayed, Worsened) 간의 만족도 평균에는 차이가 없다.
```

```python
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

# - On Time  : 출발/도착 모두 5분이내
# - Recovered :출발은 15분이상 늦었는데, 도착은 5분 이내 (만회함)
# - Delayed  : 출발도 늦고, 도착도 늦음
# - Worsened  : 출발은 제때 했는데 도착이 늦음

def classify_delay(row):
    dep=row['Departure Delay in Minutes']
    arr=row['Arrival Delay in Minutes']

    if( dep<=5 and arr<=5):
        return '1. On-Time'
    elif dep>15 and arr<=5:
        return '2. Recovered'
    elif dep>5 and arr>5:
        return '3. Delayed'
    elif dep<=5 and arr>15:
        return '4. Worsened'
    else:
        return '5. Others'
    
df['Delay_Category']=df.apply(classify_delay, axis=1)

target_groups= ['1. On-Time', '2. Recovered', '3. Delayed', '4. Worsened']
df_analysis=df[df['Delay_Category'].isin(target_groups)].copy()

# satisfied -> 1, neutral or dissatisfied -> 0
df['satisfaction_binary'] = df['satisfaction'].apply(lambda x: 1 if x == 'satisfied' else 0)

group_lists=[df_analysis[df_analysis['Delay_Category']==g]['satisfaction_binary']for g in target_groups]

f_stat, p_val= stats.f_oneway(*group_lists)

print("===Anova 결과====")
print(f"F-static:{f_stat:4f}")
print(f"P-value:{p_val:.30f}")

# 그룹별 만족도 평균 확인
print("\n=== 그룹별 만족도 평균 (비율) ===")
print(df_analysis.groupby('Delay_Category')['satisfaction_binary'].mean())

# 시각화
plt.figure(figsize=(10, 6))
sns.barplot(x='Delay_Category', y='satisfaction_binary', data=df_analysis, 
            order=target_groups, palette='viridis', errorbar=None) # errorbar=None으로 깔끔하게
plt.title('Impact of Delay Scenario on Satisfaction (The "Recovered" Paradox)')
plt.ylabel('Satisfaction Rate (0~1)')
plt.ylim(0, 0.6) # Y축 조절
plt.show()
```

```python

import matplotlib.pyplot as plt
import seaborn as sns


order_list=['1. On-Time', '2. Recovered', '3. Delayed', '4. Worsened']
fig, axes= plt.subplots(1,2,figsize=(18,6))

#1. 비행거리 분포
# 가설: recovered 그룹의 비행거리가 압도적으로 길것이다.(길어야 시간만회할수 있으니까)
sns.boxplot(x='Delay_Category', y='Flight Distance', data=df_analysis,
            order=order_list, palette='coolwarm', ax=axes[0])
axes[0].set_title('Secret 1: Flight Distance by Delay Category')
axes[0].set_ylabel('Flight Distance (Miles)')

#2. 좌석 등급비율
# 가설: recovered 그룹에 비즈니스 탑승객이 많을것이다.
ct=pd.crosstab(df_analysis['Delay_Category'], df_analysis['Class'], normalize='index')
ct=ct.reindex(order_list)

# 누적 막대그래프 그리기
ct.plot(kind='bar', stacked=True, colormap='viridis', ax=axes[1])
axes[1].set_title('Secret 2: Class Distribution by Delay Category')
axes[1].set_ylabel('Proportion')
axes[1].legend(title='Seat Class', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()

# --- [숫자로 팩트 체크] ---
print("=== 그룹별 평균 비행 거리 (Miles) ===")
print(df_analysis.groupby('Delay_Category')['Flight Distance'].mean())

print("\n=== 그룹별 비즈니스 클래스(Business) 비율 ===")
# Business 클래스인 경우만 True로 해서 평균 내면 비율이 됩니다.
print(df_analysis.groupby('Delay_Category')['Class'].apply(lambda x: (x=='Business').mean()))
```
